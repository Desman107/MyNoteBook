```python
import torch
import torch.nn as nn

# 创建输入和输出数据
x = torch.tensor([[1.0], [2.0], [3.0]], requires_grad=True)
y_true = torch.tensor([[2.0], [4.0], [6.0]])

# 定义一个简单的线性层
model = nn.Linear(1, 1)

# 定义损失函数
criterion = nn.MSELoss()

# 前向传播
y_pred = model(x)

# 计算损失
loss = criterion(y_pred, y_true)

# 反向传播
loss.backward()

# 查看梯度
print(x.grad)  # 输入x的梯度（如果x是模型参数）
print(model.weight.grad)  # 模型权重的梯度
```

```python
class TitanicSurvivalNN(nn.Module):
    def __init__(self):
        super(TitanicSurvivalNN, self).__init__()
        self.layer1 = nn.Linear(6, 128)  # First hidden layer
        self.layer2 = nn.Linear(128, 64) # Second hidden layer
        self.output_layer = nn.Linear(64, 1) # Output layer
        
        self.relu = nn.ReLU()  # ReLU activation function
        self.sigmoid = nn.Sigmoid()  # Sigmoid activation function

    def forward(self, x):
        x = self.relu(self.layer1(x))
        x = self.relu(self.layer2(x))
        x = self.sigmoid(self.output_layer(x))
        return x
```

```python
model = TitanicSurvivalNN()
matrix = torch.rand(100, 6)
actual = torch.rand(100)
```

```python
import torch.optim as optim
```

```python
model.forward(matrix)  # Forward pass
loss = criterion(model(matrix), actual)  # Calculate loss
loss.backward()  # Backward pass
```
